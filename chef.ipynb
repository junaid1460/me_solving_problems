```python
from collections import Counter
import math
import time as t

start = t.time()

x, y = 1,3
values = {'a': x, 'b':y}
flip = {'a':'b', 'b':'a'}
s = "ab" * 10 ** 
f = Counter(s)
print(x, y)
print(f)
acount = f['a'] / x
bcount = f['b'] / y
print(acount, bcount)
if acount > bcount:
    greater = 'a'
else:
    greater = 'b'
print(greater)
places = math.ceil(f[greater] / values[greater]) - 1
print("places =", places)
splits = 0
counts = 0
smaller = None
if places > 0:
    smaller = flip[greater]
    splits = math.ceil(f[smaller] / places)
    print("splits=", splits)
    counts = f[greater] %  places
    print("counts =", counts)
out = []
if counts > 0:
    temp = '{}{}'.format(greater * values[greater], smaller * splits)
    temp = temp * counts
    f[greater] -= counts *  values[greater]
    f[smaller] -= counts * splits
    out += [temp]
    print(temp)
index = greater
print("smaller =", index)
while f[index] > 0:
    c = values[index]
    if not f[index] - c >= 0:
        c = f[index]
    out += [index * c]
    f[index] -= c
    index = flip[index]
left = flip[index]
if f[left] > 0:
    temp = '*{}'.format(left * values[left])
    count = f[left] // values[left]
    out += [temp * count]
    count = f[left] - count * values[left]
    if count > 0:
        out += [ '*{}'.format(left * count)]
    
    
print('left=', left)
print(''.join(out))
print(f)
end = t.time()
print(end - start)
```

```python
from collections import Counter
x, y = 2, 3
s = 'abaaaabbaaaaabbbb'
z = Counter(s)
counta = z['a']
countb = z['b']
n = counta + countb
curA = 0
curB = 0
res = []
totalLen = 0
turn = 0
if counta < countb:
    turn = 1
else:
    turn = 0
while totalLen < n:
    if turn == 0 and counta > 0 and curA < x:
        res += ['a']
        counta -= 1
        curA += 1
        curB = 0
        totalLen += 1
        turn = 1

    elif turn == 1 and countb > 0 and curB < y:
        res += ['b']
        countb -= 1
        curB += 1
        curA = 0
        totalLen += 1
        turn = 0
    elif counta > 0 and curA < x:
        res += ['a']
        counta -= 1
        curA += 1
        curB = 0
        totalLen += 1
        turn = 1
    elif countb > 0 and curB < y:
        res += ['b']
        countb -= 1
        curB += 1
        curA = 0
        totalLen += 1
        turn = 0
    else:
        res += ['*']
        curA = curB = 0

    if counta < countb:
        turn = 1
    else:
        turn = 0



print(''.join(res))
```
